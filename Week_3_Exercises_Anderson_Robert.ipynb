##7.5 (Flattening arrays with flatten vs. ravel)
Create a 2-by-3 array containing the first six powers of 2 beginning with 2^0. Flatten the array first with method flatten, then with ravel. In each case, display the result then display the original array to show that it was unmodified.

````python
# create an array of twos
array = np.full([2,3],2)
print(array)

# transform array contents with math
for i in range(2):
  for j in range(3):
    array[i][j] = 2**(i*3+j)

# why this works:
# i * 3 + j = [0] * 3 + [0] = 1
# i * 3 + j = [0] * 3 + [1] = 1
# i * 3 + j = [0] * 3 + [2] = 2
# i * 3 + j = [1] * 3 + [0] = 3
# i * 3 + j = [1] * 3 + [1] = 4
# i * 3 + j = [1] * 3 + [2] = 5

print(array)
print(array.flatten())
print(array.ravel())
print(array)
````

## 7.14 (Horizontal and Vertical Stacking)
Create the two-dimensional arrays

```python
array1 = np.array([[0,1],[2,3]])
array2 = np.array([[4,5],[6,7]])

```

> a. Use vertical stacking to create the 4-by-2 array named array3 with array1 stacked on top of array2.
                
> b. Use horizontal stacking to create the 2-by-4 array named array4 with array2 to the right of array1.
                
> c. Use vertical stacking with two copies of array4 to create a 4-by-4 array5.
                
> d. Use horizontal stacking with two copies of array3 to create a 4-by-4 array6.

````python
# create 2d arrays
array1 = np.array([[0,1],[2,3]])
array2 = np.array([[4,5],[6,7]])

# a. Use vertical stacking to create the 4-by-2 array named array3 with array1 stacked on top of array2.
array3 = np.vstack((array1, array2))
print("Array 3: \n",array3)
print(np.shape(array3))
# b. Use horizontal stacking to create the 2-by-4 array named array4 with array2 to the right of array1.
array4 = np.hstack((array1,array2))
print("Array 4: \n",array4)
print(np.shape(array4))
# c. Use vertical stacking with two copies of array4 to create a 4-by-4 array5.
array5 = np.vstack((array4,array4)).reshape(4,4)
print("Array 5: \n",array5)
print(np.shape(array5))
# d. Use horizontal stacking with two copies of array3 to create a 4-by-4 array6.
array6 = np.hstack((array3,array3)).reshape(4,4)
print("Array 6: \n",array6)
print(np.shape(array6))
````

## 7.15 (Research and Use NumPy’s concatenate Function)
Research NumPy function concatenate, then use it to reimplement the previous exercise.

````python
# https://numpy.org/doc/stable/reference/generated/numpy.concatenate.html#numpy-concatenate
# np.concatenate((a, b), axis=0) - axis = 0 is vertical stack, axis = 1 is horizontal stack

# create 2d arrays
array1 = np.array([[0,1],[2,3]])
array2 = np.array([[4,5],[6,7]])

testoutput = np.ndarray((4,4))

# a. Use vertical stacking to create the 4-by-2 array named array3 with array1 stacked on top of array2.
array3 = np.concatenate((array1, array2), axis = 0)
print("Array 3: \n",array3)
print(np.shape(array3))
# b. Use horizontal stacking to create the 2-by-4 array named array4 with array2 to the right of array1.
array4 = np.concatenate((array1,array2), axis = 1)
print("Array 4: \n",array4)
print(np.shape(array4))
# c. Use vertical stacking with two copies of array4 to create a 4-by-4 array5.
array5 = np.concatenate((array4,array4), axis = 0).reshape(4,4)
print("Array 5: \n",array5)
print(np.shape(array5))
# d. Use horizontal stacking with two copies of array3 to create a 4-by-4 array6.
array6 = np.concatenate((array3,array3), axis = 1, out = testoutput).reshape(4,4) # testing the out= argument
print("Array 6: \n",array6)
print(np.shape(array6))
````
