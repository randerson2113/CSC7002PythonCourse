````python
import re

#write is_palindrome function - test if string forward == string reversed
def is_palindrome_string(string):
  '''test if string forward == string reversed'''
  #make the string lowercase for consistent comparisons
  string = string.lower()
  #regex keep alphanumeric characters only
  string = re.sub('[^A-Za-z0-9]+', '', string)
  #reverse the string with slicing
  string2 = string[::-1]
  #return Bool result
  return string == string2

def is_palindrome_stack(string):
  '''test if string forward == string reversed by simulating a stack'''
  #regex keep alphanumeric characters only, cast to lowercase
  string = string.lower()
  string = re.sub('[^A-Za-z0-9]+', '', string)
  #convert string to list
  string_list1 = list(string)
  #get list length
  length = len(string_list1)
  #create empty list to hold popped items
  string_list2=  []

  #loop through string_list1 and pop elements onto string_list2
  for i in range(0,length):
    letter = string_list1.pop()
    string_list2.append(letter)
  #fill up the empty list with the original string
  string_list1 = list(string)
  #print for proof of functionality - comment out
  #print(string_list1)
  #print(string_list2)
  return string_list1 == string_list2

###TEST CASES
#call function on test strings (known palindromes)
print(is_palindrome_string('Radar'))
print(is_palindrome_stack('Radar'))
print()
print(is_palindrome_string('A man, a plan, a canal, Panama'))
print(is_palindrome_stack('A man, a plan, a canal, Panama'))
print()
print(is_palindrome_string('Mr. Owl ate my metal worm'))
print(is_palindrome_stack('Mr. Owl ate my metal worm'))
print()
#call function on test strings (non-palindromes)
print(is_palindrome_string('This is a test'))
print(is_palindrome_stack('This is a test'))
print()
print(is_palindrome_string('This is not a palindrome'))
print(is_palindrome_stack('This is not a palindrome'))
print()
````

###6.11 (Analyzing the Game of Craps)
Modify the script of Fig.4.2 to play  games of craps. Use a  dictionary to keep track of the number of games won for a particular number of rolls. Similarly, use a  dictionary to keep track of the number of games lost for a particular number of rolls. As the simulation proceeds, keep updating the dictionaries.
              
                
A typical key–value pair in the  dictionary might be

  > 4: 50217
              
indicating that 50217 games were won on the 4th roll. Display a summary of the results including:
                
* a) the percentage of the total games played that were won.

* b) the percentage of the total games played that were lost.
                                
* c) the percentages of the total games played that were won or lost on a given roll (column 2 of the sample output).
                
* d) the cumulative percentage of the total games played that were won or lost up to and including a given number of rolls (column 3 of the sample output).

Output should resemble:
Rolls | % Resolved on this roll | Cumulative % of games resolved
--|--|--
1 | 30.10% | 31.105
2 | 20.80% | 50.90%
etc | etc | etc
N | x.x% | 100.0%

                

````python
#fig04_02.py
"""Simulating the dice game Craps. Modified to meet the criteria above"""
import random

def roll_dice():
    """Roll two dice and return their face values as a tuple."""
    die1 = random.randrange(1, 7)
    die2 = random.randrange(1, 7)
    return (die1, die2)  #pack die face values into a tuple

def display_dice(dice):
    """Display one roll of the two dice."""
    die1, die2 = dice  #unpack the tuple into variables die1 and die2
    print(f'Player rolled {die1} + {die2} = {sum(dice)}')

#setup the games
total_games = 50000
wins_by_roll = {}   #{roll_number: count}
losses_by_roll = {} #{roll_number: count}

for i in range(total_games):
    roll_count = 1
    die_values = roll_dice()
    #determine game status and point, based on first roll
    sum_of_dice = sum(die_values)

    if sum_of_dice in (7, 11):  #win
        game_status = 'WON'
    elif sum_of_dice in (2, 3, 12):  #lose
        game_status = 'LOST'
    else:  #remember point
        game_status = 'CONTINUE'
        my_point = sum_of_dice
    #print('Point is', my_point)

    #roll again to resolve the game, as needed
    while game_status == 'CONTINUE':
      roll_count += 1
      die_values = roll_dice()
      sum_of_dice = sum(die_values)

      if sum_of_dice == my_point:
          game_status = 'WON'
      elif sum_of_dice == 7:
          game_status = 'LOST'

    #Update dictionaries using .get() to handle new keys
    if game_status == 'WON':
        wins_by_roll[roll_count] = wins_by_roll.get(roll_count, 0) + 1
    else:
        losses_by_roll[roll_count] = losses_by_roll.get(roll_count, 0) + 1

#calculate totals of wins and losses
total_wins = sum(wins_by_roll.values())
total_losses = sum(losses_by_roll.values())

#format summary of games as table
print(f'Percentage of games won: {total_wins / total_games:.2%}')
print(f'Percentage of games lost: {total_losses / total_games:.2%}')
print(f'{"Rolls":<10}| {"% Resolved on this roll":<25}| {"Cumulative % of games resolved"}')
print('-' * 60)#table header separator

cumulative_resolved = 0.0
#Iterate through the rolls
all_rolls = sorted(set(wins_by_roll.keys()) | set(losses_by_roll.keys()))

for roll in all_rolls:
    wins = wins_by_roll.get(roll, 0)
    losses = losses_by_roll.get(roll, 0)
    resolved_this_roll = (wins + losses) / total_games
    cumulative_resolved += resolved_this_roll

    print(f'{roll:<10}| {resolved_this_roll:<25.3%}| {cumulative_resolved:.3%}')#show third decimal place for instances of 0.00%

````

###6.14 (Intro to Data Science: Dynamic Visualization of Coin Tossing)
Modify your coin-tossing simulation from Exercise 5.31 to update the bar plot dynamically as you flip the coin. Use the techniques you learned in Section 6.4.2.

````python
import random
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib import animation, rc
from IPython.display import HTML

#Tell Colab to use JavaScript for the animation
rc('animation', html='jshtml')

def update(frame_number, faces, counts):
    """dynamic visualization of dice rolls"""
    #Flip the coin 10 times per frame to speed up the visual
    for _ in range(10):
        face = random.choice(['Heads', 'Tails'])
        counts[face] += 1

    #Prepare data for plotting
    face_names = list(counts.keys())
    frequencies = list(counts.values())
    total_flips = sum(frequencies)

    #Refresh the plot
    plt.cla()
    sns.set_style('whitegrid')

    #Map x to hue and set legend=False to avoid the deprecation warning
    axes = sns.barplot(x=face_names, y=frequencies, hue=face_names,
                       palette='bright', legend=False)

    #Labels and Titles
    axes.set_title(f'Coin Tossing Simulation: {total_flips:,} Flips')
    axes.set(xlabel='Face', ylabel='Frequency')

    #Dynamically scale Y-axis with 15% headroom
    if frequencies:
        axes.set_ylim(0, max(frequencies) * 1.15)

        #Display frequency and percentage above each bar
        for bar, freq in zip(axes.patches, frequencies):
            text_x = bar.get_x() + bar.get_width() / 2.0
            text_y = bar.get_height()
            text = f'{freq:,}\n{freq / total_flips:.1%}'
            axes.text(text_x, text_y, text, ha='center', va='bottom')

#1. Setup the figure and persistent state
fig = plt.figure()
#Use a dictionary to maintain state across frames (Deitel Chapter 6)
coin_counts = {'Heads': 0, 'Tails': 0}

#Assign the animation to a variable
coin_anim = animation.FuncAnimation(
    fig, update, frames=49, interval=100,
    fargs=(['Heads', 'Tails'], coin_counts), repeat=False)

#Clean up and render for Colab
plt.close()
display(HTML(coin_anim.to_jshtml()))
````

###6.15 (Intro to Data Science: Dynamic Visualization of Rolling Two Dice)
Modify your simulation of rolling two dice from Exercise 5.32 to update the bar plot dynamically as you roll the dice. Use the techniques you learned in Section 6.4.2.

````python
import random
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib import animation, rc
from IPython.display import HTML

#Standard Colab setup for JS animations
rc('animation', html='jshtml')

def update(frame_number, rolls_per_frame, sums_dict):
    """Update function for the animation (Deitel Section 6.4.2 style)"""
    #1. Roll two dice multiple times per frame to show faster progress
    for _ in range(rolls_per_frame):
        die1 = random.randrange(1, 7)
        die2 = random.randrange(1, 7)
        dice_sum = die1 + die2
        sums_dict[dice_sum] += 1

    #2. Extract data for plotting
    sums = list(sums_dict.keys())
    frequencies = list(sums_dict.values())
    total_rolls = sum(frequencies)

    #3. Refresh the plot
    plt.cla()
    sns.set_style('whitegrid')

    #Map sums to hue and set legend=False for clean formatting
    axes = sns.barplot(x=sums, y=frequencies, hue=sums, palette='bright', legend=False)

    #4. Final Flair: Labels and Dynamic Scaling
    axes.set_title(f'Rolling Two Dice: {total_rolls:,} Total Rolls')
    axes.set(xlabel='Sum of Two Dice', ylabel='Frequency')

    if total_rolls > 0:
        axes.set_ylim(0, max(frequencies) * 1.15) #15% headroom

        #Display freq and % above each bar
        for bar, freq in zip(axes.patches, frequencies):
            text_x = bar.get_x() + bar.get_width() / 2.0
            text_y = bar.get_height()
            text = f'{freq:,}\n{freq / total_rolls:.1%}'
            axes.text(text_x, text_y, text, ha='center', va='bottom')

#Setup variables
#Pre-filling keys 2-12 so the X-axis stays consistent
dice_sums = {i: 0 for i in range(2, 13)}
fig = plt.figure('Rolling Two Dice Sums')

#Create the animation
#50 frames * 20 rolls/frame = 1,000 total rolls
dice_anim = animation.FuncAnimation(
    fig, update, frames=50, interval=100,
    fargs=(20, dice_sums), repeat=False)

plt.close() #Hide the static image
HTML(dice_anim.to_jshtml())
````
