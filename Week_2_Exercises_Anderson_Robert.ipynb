Calculate the product of a series of integers that are passed to the function , which receives an arbitrary argument list. Test your function with several calls, each with a different number of arguments.

````python
#import necessary modlues
from functools import reduce
#create a function with arbitraty arguments that calculate a series of n integers
def product(*args):
  '''calculate the product of a series of integers using reduce'''
  result = reduce(lambda x,y: x*y, *args)
  #return a series as stated in the problem
  return result

#create arbitrary series with different lengths
list1 = [1,2,3,4,5]
list2 = [1,2,3,4,5,6,7]
list3 = [1,2,2,4,3,3,6,6,8]
list4 = [2,2,2,2,2]

#print the results of passing the lists to the product function
print(product(list1))
print(product(list2))
print(product(list3))
print(product(list4))
````

###5.8(Sieve of Eratosthenes)
A prime number is an integer greater than 1 that’s evenly divisible only by itself and 1. The Sieve of Eratosthenes is an elegant, straightforward method of finding prime numbers. The process for finding all primes less than 1000 is:
              
                
    a. Create a 1000-element list [primes]
    with all elements initialized to True.

    List elements with prime indices
    (like 2, 3, 5, 7, 11, …) will remain True.

    All other list elements will eventually be set to False.
  
    b. Starting with index 2, if a given element is True
    iterate through the rest of the list and set to False every
    element in primes whose index is a multiple of the index
    for the element we’re currently processing.

    For list index 2, all elements beyond element 2
    in the list that have indices which are multiples of 2 (i.e., 4, 6, 8, 10, …, 998)
    will be set to False.
  
    c. Repeat Step (b) for the next True element.

    For list index 3 (which was initialized to True),
    all elements beyond element 3 in the list that have indices
    which are multiples of 3 (i.e., 6, 9, 12, 15, …, 999)
    will be set to False; and so on.
    A subtle observation (think about why this is true):
    The square root of 999 is 31.6, you’ll need to test and set to False only all multiples
    of 2, 3, 5, 7, 9, 11, 13, 17, 19, 23, 29 and 31.
    This will significantly improve the performance
    of your algorithm, especially if you decide
    to look for large prime numbers
  
    When this process completes, the list elements
    that are still True indicate that the index is a
    prime number. These indices can be displayed. Use
    a list of 1000 elements to determine and display
    the prime numbers less than 1000. Ignore list
    elements 0 and 1. [As you work through the book,
    you’ll discover other Python capabilities that
    will enable you to cleverly reimplement this
    exercise.]
                

````python
#Sieve of Eratosthenes
#generate empty list of prime numbers
primes = []

#set 1-1000 to true, ignore 0-1
for i in range(0,1000+1):
  primes.append(True)

textbook_list = [2,3,5,7,9,11,13,17,19,23,29,31,37]

#starting with index 2, set values whose index is a multiple of the index to false
for i in range(2,1000+1):
  #set elements divisible by textbook_list to false
  for j in textbook_list:
    if i % j == 0:
      primes[i] = False
  if i in textbook_list:
    primes[i] = True

#print the index and value of list primes
for i,j in enumerate(primes):
  if i > 1 and j == True:
    print(i, j)
````

````python
import re

#write is_palindrome function - test if string forward == string reversed
def is_palindrome_string(string):
  '''test if string forward == string reversed'''
  #make the string lowercase for consistent comparisons
  string = string.lower()
  #regex
  string = re.sub('[^A-Za-z0-9]+', '', string)
  string2 = string[::-1]
  #return Bool result
  return string == string2

def is_palindrome_stack(string):
  '''test if string forward == string reversed'''
  #regex keep alphanumeric characters only
  string = re.sub('[^A-Za-z0-9]+', '', string)
  #convert string to list
  string_list1 = list(string.lower())
  #get list length
  length = len(string_list1)
  #create empty list to hold popped items
  string_list2=  []

  #loop through string_list1 and pop elements onto string_list2
  for i in range(0,length):
    letter = string_list1.pop()
    string_list2.append(letter.lower())

  return string_list1 == string_list2


#call function on test strings (known palindromes)
print(is_palindrome_string('Radar'))
print(is_palindrome_stack('Radar'))
print()
print(is_palindrome_string('A man, a plan, a canal, Panama'))
print(stack('A man, a plan, a canal, Panama'))
print()
print(is_palindrome_string('Mr. Owl ate my metal worm'))
print(is_palindrome_string('Mr. Owl ate my metal worm'))
print()
#call function on test strings (non-palindromes)
print(is_palindrome_string('This is a test'))
print(is_palindrome_stack('This is a test'))
print()
print(is_palindrome_string('This is not a palindrome'))
print(is_palindrome_stack('This is not a palindrome'))
print()
````

###5.10(Anagrams)
An anagram of a string is another string formed by rearranging the letters in the first. Write a script that produces all possible anagrams of a given string using only techniques that you’ve seen to this point in the book. [The itertools module provides many functions, including one that produces permutations.]

````python
#anagrams
import itertools
#must pip install pyspellchecker
from spellchecker import SpellChecker

# turn off loading a built language dictionary, case sensitive on (if desired)
spell = SpellChecker(language='en', case_sensitive=False)

def anagrams(string):
  '''create all possible "valid" anagrams of a given string
      definition reduced to English words for brevity'''
  letter_list = []
  valid_word_list = []
  for i in string:
    #get each letter in the string and add to new_list
    letter_list.append(i)
  #get all permutations of letter_list
  for i in itertools.permutations(letter_list):
    #create a joined string out of i
    new_string = ''.join(i)
    #check permutations in Enlgish language dictionary, append if valid
    if new_string not in valid_word_list and new_string in spell:
      valid_word_list.append(new_string)
  #return list of anagrams
  return valid_word_list

#this is too computationally expensive for strings over 10 characters
anagram_string = 'beat'
anagrams(anagram_string)
````

````python
#create a function that receives a string and returns a list of tuples with letters and counts
import re
def summarize_letters(string):
  '''receive a string and return a list of tuples with letters and counts'''
  #create empty list
  letter_list = []
  #clean string
  cleaned_string_as_list = list(re.sub('[^A-Za-z]+', '', string).lower())
  sorted_string_as_list = sorted(cleaned_string_as_list)
  #create tuples of letter and lettercount from the cleaned string
  for i in set(sorted_string_as_list):
    letter = i # get the letter itself
    letter_count = sorted_string_as_list.count(i)
    letter_list.append((letter,letter_count))
  #return list of tuples
  print(f"Does the string \'{string}\' contain all letters? {check_alphabet(string)}")
  return letter_list

#define a separate function to check if all letters in the English alphabet exist in the string
def check_alphabet(string):
  '''check if all letters in the English alphabet exist in the string'''
  for i in range(ord('a'),ord('z')+1):
    if chr(i) not in string:
      return False
  return True

test_string = 'I am a banana!!!'
print(summarize_letters(test_string))

test_string2 = 'The quick brown fox jumps over the lazy dog'
print(summarize_letters(test_string2))
````
