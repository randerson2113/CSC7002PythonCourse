```python
#modify a column
#this will not work if any of the values are null/empty
#cars['cubicinches'] = cars['cubicinches'].astype(int)

#coerce the null data
cars['cubicinches'] = pd.to_numeric(cars['cubicinches'], errors='coerce')
cars['weightlbs'] = pd.to_numeric(cars['weightlbs'], errors='coerce')
cars.info()
```

```python
#check the total null values in the dataframe
cars.isna().sum()
```

```python
#drop null values if there is an insignificant amount
cars.dropna(inplace=True)
```

```python
#check the new totals for each column
cars.info()
cars.isnull().sum()
cars.shape
```

```python
#get specific info
cars.columns
```

```python
#apply statistical functions
# will not work for non-numeric columns
for column in cars.columns:
  print(column)
  if cars[column].dtype.name == 'object':
      print("Unique entries:",cars[column].unique())
      print("Number of each", cars[column].value_counts())
      print("most common:", cars[column].mode())
  elif cars[column].dtype.name == 'bool':
    print(cars[column].value_counts())
  else:
    print("Mean", cars[column].mean())
    print("Max", cars[column].max())
    print("Min", cars[column].min())
    print("Median", cars[column].median())
    print()


```

```python
#get a list of each column name
col = list(cars.columns)
print(col)
```

```python
#get statistics for each column
cars.describe()
```

```python
#replace null strings/objects with mode, numeric values with median
for column in cars.columns:
  if cars[column].dtype.name == 'object':
    cars[column].fillna(cars[column].mode()[0], inplace=True)
  elif cars[column].dtype.name == 'bool':
    cars[column].fillna
```

```python
#remove duplicates
cars.drop_duplicates(inplace=True)
```

```python
cars.shape
cars.describe()
```

```python
#drop specific columns
cars.drop(['year'], axis=1, inplace=True)
#axis=0 refers to rows, axis=1 refers to columns
```

```python
#delete the first record
cars = cars.drop(0, axis=0)
```

```python
cars.describe()
```

```python
#rename one or many columns
cars.rename(columns={'brand':'Make', 'cubicinches':'Cu. In.', 'hp':'HorsePower', 'weightlbs': 'Weight (lbs)'}, inplace=True)
```

```python
#filter data on specific conditions
cars[cars['HorsePower'] > 200]
filter = cars['HorsePower'] > 200
cars[filter]
```

```python
#create a filter with or condition as one statement
filter = (cars['HorsePower'] > 200) | (cars['HorsePower'] < 100)
cars[filter]
```

```python
#create a filter with and condition as one statement
filter = (cars['HorsePower'] > 200) & (cars['HorsePower'] < 300)
cars[filter]
```
